```python
google = google.drop(google.loc[google['App Name'].isna()].index)
```

```python
google.isna().sum()
```

```python
#lAST UPDATE AS RELEASE
for index, row in google.iterrows():
    if(pd.isnull(row['Released'])):
        if(pd.isnull(row['Last Updated'])):
            continue
        else:
            google.at[index, 'Released'] = google.at[index, 'Last Updated']
```

```python
google.isna().sum()
```

```python
google = google.drop(google.loc[google['Installs'].isna()].index)
```

```python
google.isna().sum()
```

```python
#I select the average of the rating to the null values
google['Rating']  = google['Rating'].astype(float)
avg = google['Rating'].mean()
google['Rating'].fillna(avg,inplace=True)
```

```python
google.isna().sum()
```

```python
google['Rating Count']  = google['Rating Count'].astype(float)
avg = google['Rating Count'].mean()
google['Rating Count'].fillna(avg,inplace=True)
```

```python
#Removed missing values
google.isna().sum()
```

```python
duplicates = google[google.duplicated()]
duplicates
```

```python
google = google.drop_duplicates()
```

```python
#Removed duplicates
duplicates = google[google.duplicated()]
duplicates
```

```python
#Create new columns
#Price type
def price_type(google):
    if google['Price'] == 0.00:
        return 'Free-to-Play'
    elif (google['Price']>0.00) & (google['Price'] <= 10.00):
        return 'Cheap'
    elif (google['Price']>10.00) & (google['Price'] <= 40.00):
        return 'Moderate'
    else:
        return 'Expensive'

google['Price_Type'] = google.apply(price_type, axis =1)
```

```python
google
```

```python
print(google.Free.value_counts())
print("Apps that have Price = 0, have Free column True")
google.loc[(google.Price==0) & (google.Free==False),'Free'] = True
print(google.Free.value_counts())
```

```python
google['Type'] = np.where(google['Free'] == True,'Free','Paid')
google.drop(['Free'],inplace=True,axis=1)
```

```python
google['Minimum Installs'] = google['Minimum Installs'].astype('int64')
google['Downloads'] = (google['Minimum Installs'] + google['Maximum Installs'])/2
google['Downloads'] = google['Downloads'].astype('int64')
google
```

```python
google = google.drop(columns=['Installs','Maximum Installs','Minimum Installs'], axis=1)
```

```python
google['Released'] = pd.to_datetime(google['Released'], errors = 'coerce')
import datetime
google['Year_Release'] = google['Released'].dt.year
```

```python
google['Rating Count'] = google['Rating Count'].astype('int64')
```

```python
google = google[(google['Released'] > '2009-12-31') & (google['Released']< '2021-01-01')]
```
